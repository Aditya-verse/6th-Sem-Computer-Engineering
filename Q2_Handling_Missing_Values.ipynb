# Handling Missing Values
This notebook contains a program for handling missing values in a dataset using mean imputation.

## Imports

```python
import pandas as pd
import numpy as np
```

## Dataset Creation

```python
# Create a sample dataset
data = { 'A': [1, 2, np.nan, 4], 'B': [np.nan, 2, 3, 4], 'C': [1, 2, 3, 4] }
df = pd.DataFrame(data)

# Display the dataset
df
```

## Handling Missing Values Using Mean Imputation

```python
# Mean imputation
df.fillna(df.mean(), inplace=True)

# Display the dataset after imputation
df
```
