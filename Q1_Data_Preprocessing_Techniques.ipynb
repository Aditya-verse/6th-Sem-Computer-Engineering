# Q1. Write a Python Program demonstrating Data Preprocessing Techniques.

## Data Preprocessing Steps:
1. Import Libraries
2. Create Dataset
3. Handle Missing Values
4. Remove Outliers
5. Encoding (Label Encoding)
6. Feature Scaling (Standardization)
7. Binning (Discretization)
8. PCA (Dimensionality Reduction)
9. Train-Test Split

```python
# Import libraries
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
```

```python
# Create dataset
data = {
    "age":[18,19,20,21,np.nan,22],
    "marks":[65,70,85,90,400,75],
    "city":["pune","mumbai","pune","delhi","mumbai","delhi"],
    "result":["pass","pass","pass","pass","fail","pass"]
}
df = pd.DataFrame(data)
print("Original Dataset:")
print(df)
```

```python
# Handling missing values
print("\n--- Handling Missing Values ---")
df["age"].fillna(df["age"].mean(), inplace=True)
print("After filling missing age values:")
print(df)
```

```python
# Remove outliers
print("\n--- Removing Outliers ---")
df = df[df["marks"] < 100]
print("After removing outliers (marks >= 100):")
print(df)
```

```python
# Encoding (Label Encoding)
print("\n--- Label Encoding ---")
le = LabelEncoder()
df["city"] = le.fit_transform(df["city"])
df["result"] = le.fit_transform(df["result"])
print("After encoding categorical variables:")
print(df)
```

```python
# Feature scaling (Standardization)
print("\n--- Feature Scaling ---")
scaler = StandardScaler()
df[["age","marks"]] = scaler.fit_transform(df[["age","marks"]])
print("After scaling numerical features:")
print(df)
```

```python
# Binning (Discretization)
print("\n--- Binning ---")
df["age_group"] = pd.cut(df["age"], bins=3, labels=["young","middle","old"])
print("After binning age into groups:")
print(df)
```

```python
# PCA (Dimensionality Reduction)
print("\n--- PCA (Dimensionality Reduction) ---")
pca = PCA(n_components=2)
reduced = pca.fit_transform(df[["age","marks","city"]])
print("After PCA (reduced to 2 components):")
print(reduced)
```

```python
# Train-Test Split
print("\n--- Train-Test Split ---")
X = df.drop("result", axis=1)
y = df["result"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
print("Training set size:", X_train.shape)
print("Testing set size:", X_test.shape)
print("\nData preprocessing completed successfully!")
```
